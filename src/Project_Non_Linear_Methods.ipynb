```r
data <- read.csv("../data/epfl-bio-322a/training_data.csv")
test_set <- read.csv("../data/epfl-bio-322a/test_data.csv")
```

```r
data <- data[, -3] #SWEET OR SOUR is not in the test set
data <- na.omit(data)
data.response <- as.integer(data$VALENCE.PLEASANTNESS)
data.predictors <- data[, -2]
data.predictors$Intensity <- as.numeric(data.predictors$Intensity)
data2 <- data.predictors[, apply(data.predictors, 2, function(v) var(v)!=0)]
```

```r
set.seed(127)
i <- sample(nrow(data), 3*nrow(data)/4)
datatrain <- as.matrix(data.predictors[i,])
datatest <- as.matrix(data.predictors[-i,])
ytrain <- data.response[i]
ytest <- data.response[-i]
```

```r
set.seed(135)
i <- sample(nrow(reduced_data), 3*nrow(reduced_data)/4)
aftertrain <- reduced_data[i,]
aftertest <- reduced_data[-i,]
afterYtrain <- data.response[i]
afterYtest <- data.response[-i]
```

# NEURAL NETWORKS

```r
library(tensorflow)
use_session_with_seed(24) 
```

```r
reduced_data <- as.matrix(reduced_data)
aftertrain <- as.matrix(aftertrain) ; aftertest <- as.matrix(aftertest)
testset <- as.matrix(testset)
```

```r
#defining scaling functions
get.scale <- function(scaled) {
    if ("scaled:center" %in% names(attributes(scaled))) {
        center <- attr(scaled, "scaled:center")
    } else {
        center <- rep(0, ncol(scaled))
    }
    if ("scaled:scale" %in% names(attributes(scaled))) {
        list(center, attr(scaled, "scaled:scale"))
    } else {
        list(center, rep(1., length(center)))
    }
}
scaleXas <- function(x, scaled) {
    s <- get.scale(scaled)
    centered <- sweep(x, 2, s[[1]])
    sweep(centered, 2, s[[2]], FUN = "/")
}
scaleYas <- function(y, scaled) {
    s <- get.scale(scaled)
    (y - s[[1]])/s[[2]]
}
unscaleY <- function(y, scaled) {
    s <- get.scale(scaled)
    y * s[[2]] + s[[1]]
}
```

```r
aftertrain.scaled <- scale(aftertrain, center = T, scale = T)
afterYtrain.scaled <- scale(afterYtrain, center = T, scale = T)
aftertest.scaled <- scaleXas(aftertest, aftertrain.scaled)
reduced_data.scaled <- scaleXas(reduced_data, aftertrain.scaled)
testset.scaled <- scaleXas(testset, aftertrain.scaled)
afterYtest.scaled <- scaleYas(afterYtest, afterYtrain.scaled)
data.response.scaled <- scaleYas(data.response, afterYtrain.scaled)
```

First, we fitted a neural network with dense layers with relu as activation funcion, because it is made for regression. We also introduced dropout layers.

```r
nn <- keras_model_sequential()
nn <- nn %>%
      layer_dense(500, activation = "relu", input_shape = ncol(aftertrain.scaled)) %>%
      layer_dense(200, activation = "relu") %>%
      layer_dropout(rate = .3) %>%
      layer_dense(300, activation = "relu") %>%
      layer_dropout(rate = .2) %>%
      layer_dense(100, activation = "relu") %>%
      layer_dense(1)
nn %>% compile(optimizer = "adam", loss = "mean_squared_error", metrics = list("mean_absolute_error"))
history <- nn %>% fit(aftertrain.scaled,
                      afterYtrain.scaled,
                      verbose = 0,
                      batch_size = length(afterYtrain.scaled),
                      validation_data = list(aftertest.scaled, afterYtest.scaled),
                      epochs = 500)
nn.pred <- predict(nn, aftertest.scaled)
mean((unscaleY(nn.pred, afterYtrain.scaled) - afterYtest)^2)
```

```r
plot(history)
```

There is an overfit as validation loss and error have a U shape and is higher than training loss and error.

Now we introduce regularization in our dense layers, with L1 regularization.

```r
#with regularization
nn <- keras_model_sequential()
nn <- nn %>%
      layer_dense(500, activation = "relu", kernel_regularizer= regularizer_l1(0.1), bias_regularizer=regularizer_l1(0.1), input_shape = ncol(aftertrain.scaled)) %>%
      layer_dropout(rate = 0.4) %>%
      layer_dense(200, activation = "relu", kernel_regularizer= regularizer_l1(0.1), bias_regularizer=regularizer_l1(0.1)) %>%
      layer_dropout(rate = .3) %>%
      layer_dense(300, activation = "relu", kernel_regularizer= regularizer_l1(0.1), bias_regularizer=regularizer_l1(0.1)) %>%
      layer_dropout(rate = .2) %>%
      layer_dense(100, activation = "relu", kernel_regularizer= regularizer_l1(0.1), bias_regularizer=regularizer_l1(0.1)) %>%
      layer_dense(1)

nn %>% compile(optimizer = "adam", loss = "mean_squared_error", metrics = list("mean_absolute_error"))
history <- nn %>% fit(aftertrain.scaled,
                      afterYtrain.scaled,
                      verbose = 0,
                      batch_size = length(afterYtrain.scaled),
                      validation_data = list(aftertest.scaled, afterYtest.scaled),
                      epochs = 500)
nn.pred <- predict(nn, aftertest.scaled)
mean((unscaleY(nn.pred, afterYtrain.scaled) - afterYtest)^2)
```

```r
plot(history)
```

We see there is no more the problem of overfitting for the loss (mean squared error). We see a strange behaviour for the metric mean absolute error : validation MAE is lower.

Instead of regularization, we tried early stopping, introducing callbacks:

```r
#with early stopping
nn <- keras_model_sequential()
nn <- nn %>%
      layer_dense(500, activation = "relu", input_shape = ncol(aftertrain.scaled)) %>%
      layer_dropout(rate = 0.4) %>%
      layer_dense(200, activation = "relu") %>%
      layer_dropout(rate = .3) %>%
      layer_dense(300, activation = "relu",) %>%
      layer_dropout(rate = .2) %>%
      layer_dense(100, activation = "relu") %>%
      layer_dense(1)
callback <- callback_early_stopping(monitor = "val_loss")
nn %>% compile(optimizer = "adam", loss = "mean_squared_error", metrics = list("mean_absolute_error"))
history <- nn %>% fit(aftertrain.scaled,
                      afterYtrain.scaled,
                      verbose = 0,
                      batch_size = length(afterYtrain.scaled),
                      validation_data = list(aftertest.scaled, afterYtest.scaled),
                      callbacks = callback,
                      epochs = 500)
nn.pred <- predict(nn, aftertest.scaled)
mean((unscaleY(nn.pred, afterYtrain.scaled) - afterYtest)^2)
```

```r
history$params$epochs <- length(history$metrics$loss)
plot(history)
```

Again, we observe a strange validation loss and MAE behaviour: it may be because of dropouts. Early stopped neural network MSE is higher than Regularized neural network MSE.

The best Neural Network model includes Regularization.

# BOOSTING

We first tried a model with boosting.
We tried different cross validations and parameters tuning.

```r
#Manual tuning
pows <- seq(-3, -1, by = 0.2)
lambdas <- 10^pows
depths <- seq(1, 9, 1)
train.err <- matrix(nrow = length(depths), ncol = length(lambdas))
test.err <- matrix(nrow = length(depths), ncol = length(lambdas))

for (k in 1:length(depths)){
    for (i in 1:length(lambdas)) {
        boost.valence <- xgboost(as.matrix(aftertrain),
                                label= afterYtrain,
                                eta=lambdas[i],
                                max.depth=depths[k],
                                objective="reg:squarederror",
                                nrounds=10000,
                                early_stopping_rounds = 5, #if error is not increased after 5 steps, boosting will stop.
                                verbose = 0)
        pred.train <- predict(boost.valence, as.matrix(aftertrain))
        pred.test <- predict(boost.valence, as.matrix(aftertest))
        #at line k there will be the kth depth and at column i will be the ith lambda
        train.err[k,i] <- mean((pred.train - afterYtrain)^2)
        test.err[k,i] <- mean((pred.test - afterYtest)^2)
    }
}
```

```r
for (k in 1:length(depths)){
    par(mfcol=c(1,2), cex=1.1)
    plot(log(lambdas), log(train.err[k,]), col = "red", type="b", xlab="log Shrinkage values", ylab="log TRAIN MSE")
    plot(log(lambdas), log(test.err[k,]), type="b", col = "blue", xlab="log Shrinkage values", ylab="log TEST MSE")
    points(log(lambdas[which.min(test.err[k,])]), log(min(test.err[k,])), col="red", pch=20, cex=2)
    cat("tree depth of : " , k)
}
```

```r
which.min(test.err)
```

the 19th value corresponds to column 3 and line 1: so for a tree depth of 1 and a lambda of 10^(-2.6) which is 0.002512.

```r
set.seed(24)
#reproduce the best boosted model with depth and lambda found above.
#fitting model on splitted train data
boostingtest <- xgboost(as.matrix(aftertrain), label = afterYtrain,
                 eta = 10^(-2.6),
                 max.depth = 1,
                 nrounds = 10000,
                 early_stopping_rounds = 5,
                 verbose = 0,   
                 objective = "reg:squarederror") 
#prediction on splitted test data and MSE computation
testmse <- mean((predict(boostingtest, as.matrix(aftertest)) - afterYtest)^2)
testmse
```

```r
#prediction on complete reduced data
boosting <- xgboost(as.matrix(reduced_data), label = data.response,
                 eta = 10^(-2.6),
                 max.depth = 1,
                 nrounds = 10000,
                 early_stopping_rounds = 5,
                 verbose = 0,   
                 objective = "reg:squarederror") 
responses_predicted <- predict(boosting, as.matrix(testset))
write.csv(data.frame(Id = 1:length(responses_predicted), VALENCE.PLEASANTNESS = responses_predicted), 'BoostedManual.csv', row.names = F)
```

```r
xgb.plot.importance(xgb.importance(model = boosting), top_n = 10)
```

```r
#cross validation using xgb.cv
for (depths in c(1, 3, 5, 7, 9)){
    cat("depth = " , depths, "\n")
    for (etas in c(0.001, 0.01, 0.1)){
        cat("\n" ,"eta = " , etas, "\n")
        parametres <- list(booster = "gbtree", objective = "reg:squarederror", eta=etas, max_depth= depths , eval_metric = "rmse")
        xgbcv <- xgb.cv(params = parametres, data = xgb.DMatrix(as.matrix(aftertrain), label =afterYtrain), nrounds = 10000, nfold = 5, showsd = T, stratified = T, print_every_n = 100, early_stopping_rounds=T)
    }
}
```

The best test rmse (which is validation rsme) is 22.620559 obtained for max depth of 3 and eta of 0.1.

# RANDOM FORESTS

```r
set.seed(135)
#we first tried manually to tune hyperparameters
mInit <- c(29, 203, 406, 609, 812, 1015, 1218, 1421, 1624)
train.mse <- matrix(nrow = length(mInit), ncol = 700)
test.mse <- matrix(nrow = length(mInit), ncol = 700)

for (i in 1:length(mInit)){
  rf.valence <- randomForest(x = aftertrain, y = afterYtrain , xtest = aftertest, ytest = afterYtest, mtry = mInit[i], ntree=700,importance = TRUE, na.action = na.omit)
  train.mse[i,] <- rf.valence$mse      
  test.mse[i,] <- rf.valence$test$mse
}
```

```r
dim(train.mse)
dim(test.mse)
```

```r
whmin <- 0; valmin <- 1000; mmin <- 0
for (i in 1:length(mInit)){
  if(min(test.mse[i,]) < valmin){
    valmin <- min(test.mse[i,]) 
    whmin <- which.min(test.mse[i,])
    mmin <- mInit[i]
  }
}
```

```r
valmin
whmin
mmin
```

Best random forest model obtained manually is for mInit = 1218 and nbtrees = 71, with an MSE of 424. 

```r
for (i in 1:length(mInit)){
    par(mfcol=c(1,2), cex=1.1)
    plot(1:700, log(train.mse[i,]), col = "red", type="b", xlab="number of trees", ylab="log TRAIN MSE")
    plot(1:700, log(test.mse[i,]), type="b", col = "blue", xlab="number of trees", ylab="log TEST MSE")
    points(which.min(log(test.mse[i,])), min(log(test.mse[i,])), col="red", pch=20, cex=2)
    cat("m initial of : " , mInit[i])
}
```

```r
#Fitting model on train splitted set and MSE computation:
set.seed(24)
pleasantness.rf1 <- randomForest(afterYtrain ~ ., data = aftertrain, mtry = 1218, ntree=71,importance = TRUE, na.action = na.omit)
predictionrf1 <- predict(pleasantness.rf1, aftertest, predict.all = T)
mean((predictionrf$aggregate - afterYtest)^2)
```

```r
#we then tuned random forests parameters with the tuneRF function
for (nbtrees in c(200, 300, 400, 500, 600, 700, 800)){
    cat("nb trees = " , nbtrees, "\n")
    tuneRF(aftertrain, afterYtrain, ntreeTry=nbtrees, stepFactor=2, improve=0.05,trace=TRUE, doBest=TRUE)
}
```
