```r
# play around with settings of xgboost - eXtreme Gradient Boosting (Tree) library
# https://github.com/tqchen/xgboost/wiki/Parameters
# max.depth - maximum depth of the tree
# nrounds - the max number of iterations
```

```r
library(Metrics)
smallestError <- 100
i<-0
for (depth in seq(1,10,1)) {
        for (rounds in seq(1,20,1)) {
            for(eta in seq(0.0001,0.01,0.01)) {
                print(i)
                i<-i+1
                # train
                bst <- xgboost(data = as.matrix(aftertrain),
                               label = afterYtrain, neta=eta,
                               max.depth=depth, nround=rounds,
                               objective = "reg:squarederror", verbose=0)
                gc()
               
                # predict
                predictions <- predict(bst, as.matrix(aftertest), outputmargin=TRUE)
                aftertest<-unlist(aftertest, recursive = TRUE, use.names = TRUE)
                err <-  rmse(as.numeric(aftertest), as.numeric(predictions))
               
                if (err < smallestError) {
                        print(i)
                        smallestError = err
                        print(paste(depth,rounds,err))
                }
            }
        }
} 
```

```r
cv <- 30
cvDivider <- floor(nrow(aftertrain) / (cv+1))

smallestError <- 100
for (depth in seq(1,10,1)) { 
        for (rounds in seq(1,20,1)) {
            for(eta in seq(0.0001,0.01,0.01)) {
                totalError <- c()
                indexCount <- 1
                for (cv in seq(1:cv)) {
                        # assign chunk to data test
                        dataTestIndex <- c((cv * cvDivider):(cv * cvDivider + cvDivider))
                        dataTest <- aftertrain[dataTestIndex,]
                        # everything else to train
                        dataTrain <- aftertrain[-dataTestIndex,]
                       
                        bst <- xgboost(data = as.matrix(aftertrain),
                                       label = aftertrain,
                                       max.depth=depth, nround=rounds,
                                       objective = "reg:squarederror")
                        predictions <- predict(bst, as.matrix(aftertest), outputmargin=TRUE)
                       
                        err <- rmse(as.numeric(aftertest), as.numeric(predictions))
                        totalError <- c(totalError, err)
                }
                if (mean(totalError) < smallestError) {
                        smallestError = mean(totalError)
                        print(paste(depth,rounds,smallestError))
                } 
            }
        }
} 
```

```r
# Helper packages
library(dplyr)
library(tidymodels)
library(vip)         # for variable importance plots
```

```r
tune_spec <- 
  decision_tree(
    cost_complexity = tune(),
    tree_depth = tune()
  ) %>% 
  set_engine("rpart") %>% 
  set_mode("regression")

tune_spec
```

```r
tree_grid <- grid_regular(cost_complexity(),
                          tree_depth(),
                          levels = 5)
#The function grid_regular() is from the dials package. It chooses sensible values to try for each hyperparameter; here, we asked for 5 of each. Since we have two to tune, grid_regular() returns 5 
#×5 = 25 different possible tuning combinations to try in a tidy tibble format.
```

```r
tree_grid
```

```r
tree_grid %>% 
  count(tree_depth)
```

```r
set.seed(234)
cell_folds <- vfold_cv(trainCV)
```

```r
set.seed(345)

tree_wf <- workflow() %>%
  add_model(tune_spec) %>%
  add_formula(afterYtrain ~ .)

tree_res <- 
  tree_wf %>% 
  tune_grid(
    resamples = cell_folds,
    grid = tree_grid
    )

tree_res
```

```r
tree_res %>% 
  collect_metrics()
```

```r
tree_res %>%
  collect_metrics() %>%
  mutate(tree_depth = factor(tree_depth)) %>%
  ggplot(aes(cost_complexity, mean, color = tree_depth)) +
  geom_line(size = 1.5, alpha = 0.6) +
  geom_point(size = 2) +
  facet_wrap(~ .metric, scales = "free", nrow = 2) +
  scale_x_log10(labels = scales::label_number()) +
  scale_color_viridis_d(option = "plasma", begin = .9, end = 0)
```

```r
tree_res %>%
  show_best("rmse")
```

```r
best_tree <- tree_res %>%
  select_best("rmse")

best_tree
```

```r
final_wf <- 
  tree_wf %>% 
  finalize_workflow(best_tree)

final_wf
```
